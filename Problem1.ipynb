```python
#PROBLEM 1

import numpy as np
from matplotlib import pylab as pt

m=512;
n=512;
s =80;
def modn(x, M):
    if (0<=x<M):
        return x
    else:
        x = ((x%M)+M)%M
        return x

img = np.fromfile('lena.bin',dtype='uint8',count=-1)
img = img.reshape(m,n)
#print('Display image in python ...')
f = pt.figure()
f.add_subplot(2,2, 1)
pt.title("Original Image")
pt.imshow(img)
# Do something ( e.g., invert image)
#initialization
imgShift = np.zeros((m,n),dtype='uint8')
for k in range(m):
    for l in range(n):
       
        imgShift[k][l] =  img[modn(k+s,m)][modn(l+s,n)]
f.add_subplot(2,2, 2)
pt.title("Shifted Image")
pt.imshow(imgShift)               
```
