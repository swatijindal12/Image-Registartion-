```python
import matplotlib.image as img 
import numpy as np
from matplotlib import pylab as pt


image = img.imread("result.jpg"); 
f = pt.figure()
f.add_subplot(2,2, 1)
pt.imshow(image)


width, height = image.shape[:2]; # determining the length of original image 
print(width,height)

# xAxis and yAxis are new width and height of image   

xAxis = int(width * 3); 
yAxis = int(height * 3); 
  
# calculating the scaling factor  
xScale = xAxis/(width);     # using this calculate zoomin factor 
yScale = yAxis/(height); 
print('\n')
print(xAxis, yAxis)    
print(xScale, yScale)
#print(image[100][100])
#print(image[100][101])
# using numpy taking a matrix of xAxis width and yAxis height
ZoomInImage = np.zeros([xAxis, yAxis]); 
  
for i in range(xAxis-1): 
    for j in range(yAxis-1): 
        ZoomInImage[i , j ]= image[int(i / xScale), 
                                 int(j / yScale)] 
  
#Save the image after scaling 
#print("-------------------")
#print(ZoomInImage[100][100])
#print(ZoomInImage[100][101])
#print(ZoomInImage[101][101])
#print(ZoomInImage[101][100])

img.imsave('ZoomInImg.jpg', ZoomInImage); 
f.add_subplot(2,2, 2)
pt.imshow(ZoomInImage)
```
