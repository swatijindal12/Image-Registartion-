```python
import matplotlib.image as img 
import numpy as np 
from matplotlib import pylab as pty
#row=int(input("Enter number of rows:"))
#col=int(input("Enter numbrr of columns:"))

#entries= list(map(int,input().split()))

#first_mat=np.array(entries).reshape(row,col)
#print("First matrix is: ")
#print(first_mat)
m = img.imread("result.jpg");
#print(m)
#input()
f = pty.figure()
f.add_subplot(2,2, 1)
pty.imshow(m)
row, col = m.shape[:2]; # determining the length of original image
print(row,col)
z=2;
xNew = int(row / z); 
yNew = int(col / z); 
print(xNew, yNew)
print("original image------", m[0][0],m[0][1],m[1][0],m[1][1])
newImage = np.zeros([xNew, yNew]);
#print(newImage)
for i in range(xNew-1): 
    for j in range(yNew-1):

        for k in range(z):
            for l in range(z):
                newImage[i][j]+=  m[(i)*z +k , (j)*z +l]
                
              
#newImage = sum
print("-------------")
#print(newImage)
#print("new image------", newImage)
newImage/= (z*z)
print("------------")
newImage = newImage.astype(int)  # to convert matrix into int
print(newImage)
f.add_subplot(2,2, 2)
pty.imshow(newImage)
img.imsave('zoomImage.jpg', newImage)
```
