```python
#PROBLEM2

import numpy as np
from numpy.fft import fft2, ifft2, fftshift
from matplotlib import pylab as pt

m=512;
n=512;

img = np.fromfile('lena.bin',dtype='uint8',count=-1)
img = img.reshape(m,n)

#Take FFT of Lena Image
fourier = fft2(img)
#print(fourier)


#Real Part of Original Image
Center_Fourier = fftshift(fourier)
img_real = np.log(abs((np.real(Center_Fourier))))
#print(img_real)


#Imaginary Part Of Original Image

img_imaginary = np.log(abs(np.imag(Center_Fourier)))
#print("\n")
#print(img_imaginary)

f = pt.figure()
f.add_subplot(2,2, 1)
pt.title("Real Image")
pt.imshow(img_real)

f.add_subplot(2,2, 2)
pt.title("Imaginary Image")
pt.imshow(img_imaginary)


s =80;
def modn(x, M):
    if (0<=x<M):
        return x
    else:
        x = ((x%M)+M)%M
        return x



imgShift = np.zeros((m,n),dtype='uint8')
for k in range(m):
    for l in range(n):
        imgShift[k][l] =  img[modn(k+s,m)][modn(l+s,n)]
        
        
# Take fft of Shifted Image
shift_Fourier = fft2(imgShift)


#Real Part of Shifted Image
shift_Center_Fourier = fftshift(shift_Fourier)
shiftimg_real = np.log(abs(np.real(shift_Center_Fourier)))

#print(shiftimg_real)
f.add_subplot(2,2, 3)
pt.title("Real Shifted Image")
pt.imshow(shiftimg_real)
#pt.imshow(ifft_shift)

#imaginary Part of Shifted image

shiftimg_imaginary = np.log(abs(np.imag(shift_Center_Fourier)))
#print(shiftimg_imaginary)
f.add_subplot(2,2, 4)
pt.title("Imaginary Shifted Image")
pt.imshow(shiftimg_imaginary)
```
